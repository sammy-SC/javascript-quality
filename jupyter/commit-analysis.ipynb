```python
import pandas as pd
import numpy as np
from datetime import datetime
%matplotlib inline
import matplotlib.pyplot as plt

major_releases = [
    "2016-04-26",
    "2015-10-29",
    "2015-09-08",
    "2015-08-04",
    "2015-05-04",
    "2015-01-14"
]

day_commits = []
last_day_string = None
commit_labels = []

with open("../data/commits-express.csv") as f:
    for (i, line) in enumerate(f):
        date_object = datetime.strptime(line, '%Y-%m-%d %H:%M:%S\n')
        day_key = date_object.strftime('%Y-%m')
        if last_day_string is None or day_key != last_day_string:
            day_commits.append(1)
            commit_labels.append(day_key)
        elif day_key == last_day_string:
            day_commits[-1] += 1
        last_day_string = day_key
        

plt.figure(figsize=(55, 15))
plt.plot(day_commits)

plt.ylabel('Number of commits')
plt.xticks(range(len(day_commits)), commit_labels, rotation='vertical')
plt.grid(True)
plt.show()
    
```
